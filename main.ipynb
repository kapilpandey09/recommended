```python
import ast
def conver(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import ast
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter +=1
        else:
            break
    return L
```

```python
movies["cast"] = movies.cast.apply(convert)
```

```python
movies
```

```python
movies.crew[0]
```

```python
def crew_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            L.append(i['name'])
            break
    return L
```

```python
movies["crew"].apply(crew_director)
```

```python
movies["crew"] = movies["crew"].apply(crew_director)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
# print(cv.fit_transform(new_df['tags']))
import matplotlib.pyplot as plt

# Example: Plotting histogram of non-zero values
plt.figure(figsize=(10, 6))
plt.hist(cv.fit_transform(new_df['tags']).data, bins=50, alpha=0.75)
plt.title('Histogram of Non-Zero Values in Sparse Matrix')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
# problem
# [love, loved, loving] this are same mean
# so convert a single word ['love']
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
def recommend(movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        # print(i[0])
        print(new_df.iloc[i[0]].title)
        # print(i[0])
    
```
